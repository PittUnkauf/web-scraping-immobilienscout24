```python
# Webcrawl real rent real estate data from immobilienscout 
import requests 
from bs4 import BeautifulSoup
from datetime import date

date = date.today()

r = requests.get("https://www.immobilienscout24.de/Suche/de/baden-wuerttemberg/stuttgart/wohnung-mieten")
c = r.content

soup = BeautifulSoup(c, "html.parser")

page_nr = soup.find_all("div", {"class":"select-input-wrapper"})

page_nr[0].text

number_string = "1"
i = 1
max_page_nr = 1
while number_string != page_nr[0].text:
    i = i+1
    number_string = number_string+str(i)
    max_page_nr = i

print(max_page_nr)

l = []
base_url = "https://www.immobilienscout24.de/Suche/de/baden-wuerttemberg/stuttgart/wohnung-mieten?pagenumber="


for page in range (1, max_page_nr+1):
    print(base_url + str(page))
    r = requests.get(base_url + str(page))
    c = r.content
    
    soup = BeautifulSoup(c, "html.parser")
    
    all = soup.find_all("div", {"class":"result-list-entry__data"})

    for property in all:
        d = {}
        
        try:
            d["Kaltmiete"] = property.find_all("dd", {"class":"font-nowrap font-highlight font-tabular"})[0].text.replace(" €", "").replace(".", "")
            d["Wohnfläche"] = property.find_all("dd", {"class":"font-nowrap font-highlight font-tabular"})[1].text.replace(" m²", "") 
            zimmer = property.find_all("dd", {"class":"font-nowrap font-highlight font-tabular"})[2].text
            d["Zimmer"] = zimmer[:zimmer.find(" ")]
            d["Adresse"] = property.find_all("div", {"class":"result-list-entry__address"})[0].text
            
            l.append(d)
        except:
            pass
            
```

```python
l
```

```python
# data preparation
import pandas as pd

df = pd.DataFrame(l)

# split column 'Adresse'
foo = lambda x: pd.Series([i for i in reversed(x.split(','))])
df[['Stadt','Bezirk/Stadtteil', 'Straße', 'Adresszusatz']] = df['Adresse'].apply(foo)
df[df.columns] = df.apply(lambda x: x.str.strip())
del df["Adresse"]
df.head()
```

```python
# import Matching-Tabelle (Bezirk-Stadtteil)
bezirk = pd.read_csv("Stuttgart_Bezirk_Stadtteil.csv", sep=";", encoding='unicode_escape')
bezirk.head(5)
```

```python
# Data manipulation
df = df[(df.Stadt=="Stuttgart") & (df.Zimmer!="1")].merge(bezirk, how='left', on="Bezirk/Stadtteil")

df = df.drop(['Stadt', 'Bezirk/Stadtteil', 'Straße', 'Adresszusatz'], axis=1)

df['Kaltmiete'] = df['Kaltmiete'].str.replace(',', '.').astype(float)
df['Wohnfläche'] = df['Wohnfläche'].str.replace(',', '.').astype(float)
df['Zimmer'] = df['Zimmer'].str.replace(',', '.').astype(float)

df.head()
```

```python
# Aggregation
df = df.groupby('Bezirk').agg({'Kaltmiete':'mean', 'Wohnfläche':'mean', 'Zimmer':'median'})
df = df.reset_index()
df[['Kaltmiete', 'Wohnfläche']] = df[['Kaltmiete', 'Wohnfläche']].round(0)
df.head()
```

```python
import geopandas
import folium
import fiona
```

```python
# import Ploygon-Layer Stuttgart
layers = fiona.listlayers("KLGL_Stadtplanbasis.gdb")

for layer in layers:
    stuttgart = geopandas.read_file("KLGL_Stadtplanbasis.gdb",layer=layer)

stuttgart.head(5)
```

```python
# aggregate to Bezirke
stuttgart = stuttgart[['STADTBEZIRKNAME', 'geometry']]
sttg_bezirk = stuttgart.dissolve(by='STADTBEZIRKNAME').reset_index()
sttg_bezirk
```

```python
# merge data and geodata
df = pd.merge(sttg_bezirk, df, how='left', left_on="STADTBEZIRKNAME", right_on="Bezirk")
df
```

```python
# display map
m = folium.Map(tiles='cartodbpositron', location=[48.78232, 9.17702], width=690, height=500)

folium.Choropleth(
    geo_data=sttg_bezirk,
    name='choropleth',
    data=df,
    columns=['STADTBEZIRKNAME', 'Kaltmiete'],
    key_on="feature.properties.STADTBEZIRKNAME",
    fill_color='BuPu',
    nan_fill_color="white",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='durschnittliche Kaltmieten in EUR (Stand:'+str(date)+')',
).add_to(m)

style_function = lambda x: {'fillColor': '#ffffff', 
                            'color':'#000000', 
                            'fillOpacity': 0.1, 
                            'weight': 0.1}
highlight_function = lambda x: {'fillColor': '#000000', 
                                'color':'#000000', 
                                'fillOpacity': 0.50, 
                                'weight': 0.1}

tp = folium.features.GeoJson(
    df, 
    style_function=style_function, 
    control=False,
    highlight_function=highlight_function, 

    tooltip=folium.features.GeoJsonTooltip(
        fields=['STADTBEZIRKNAME', 'Kaltmiete', 'Wohnfläche', 'Zimmer'],
        aliases=['Bezirk: ','durchschnittl. Kaltmiete in EUR: ', 'durchschnittl. Wohnfläche in m²: ', 'Zimmer-Median'],
        style=("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 10px;")
    )
)

m.add_child(tp)
m.keep_in_front(tp)
folium.LayerControl().add_to(m)

m.save("Stuttgart_Mieten.html")
m
```
